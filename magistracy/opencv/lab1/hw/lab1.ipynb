```python
def replace(image):
    for i in range(100, 250):
        for j in range(300, 450):
            buf = np.copy(image[i, j])
            image[i, j] = np.copy(image[i - 70, j + 300])
            image[i - 70, j + 300] = np.copy(buf)
    return image

img = cv2.imread('starry_night.jpg')
plt.imshow(replace(bgr_to_rgb(img)))
```
