```python
# Просим OpenCV загрузить изображение из файла
# OpenCV по умолчанию загружает изображение в BGR
img_bgr = cv.imread("assets/starry_night.jpg")

# Конвертируем изображение из BGR в RGB
img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

# План действий по трансформированию изображения из BGR в RGB:
# - разделяем матрицу изображения на три канала (например, помещаем в отдельные переменные)
# - комбинируем (располагаем) эти каналы в порядке RGB

plt.imshow(img_rgb)
```
